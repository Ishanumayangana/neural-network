```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# load dataset
data = pd.read_csv("Reduce overfitting dataset.csv", header=None)
data = data.values
```

```python
data[:1]
```

```python
# divide dataset into x and y
x = data[:,0:60].astype(float)
y = data[:,60]
```

```python
# encode labels
encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
```

```python
# scale features
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
# divide dataset into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, x, test_size=0.2, random_state=42)
```

```python
tf.random.set_seed(42)

'''
# create a model without dropout layers
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30, activation='relu', input_shape=(60,)),
    tf.keras.layers.Dense(60, activation='relu'),
    tf.keras.layers.Dense(60, activation='relu'),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])'''

 # create a model with droput layers
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30, activation='relu', input_shape=(60,)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation='sigmoid')
 ])
```

```python
# compile the model
model.compile(loss = tf.keras.losses.BinaryCrossentropy(),
             optimizer = tf.keras.optimizers.Adam(learning_rate=0.005),
             metrics = ["accuracy"])
```

```python
# train the model
epoch_number = 100
history = model.fit(x_train, y_train, epochs=epoch_number)
```

```python
# evaluate the model
model.evaluate(x_test, y_test)
```
